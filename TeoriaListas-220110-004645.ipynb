<font size="7"><span style="color: #33D6CC;">&iquest;Qu&eacute; es una lista?</span></font>

<font size="3"><span style="color: #5E62D1;">Una lista <strong>(list)</strong> en python es <strong>un conjunto, secuencia o coleccin de elementos.</strong> </span></font>

<font size="3"><span style="color: #5E62D1;">Los elementos que forman parte de una lista pueden ser de diferente tipo:</span></font>

<ul>
<font size="3"><li><span style="color: #5E62D1;">Enteros</span></li></font>
<font size="3"><li><span style="color: #5E62D1;">Flotantes</span></li></font>
<font size="3"><li><span style="color: #5E62D1;">Cadenas (strings)</span></li></font>
<font size="3"><li><span style="color: #5E62D1;">Arreglos (arrays)</span></li></font>
<font size="3"><li><span style="color: #5E62D1;">Inclusive listas y ms</span></li></font>
</ul>

<font size="3"><span style="color: #5E62D1;">La estructura de una lista consiste en escribir los elementos de dicha lista dentro de corchetes <strong>[ ].</strong></span></font>

<font size="3"><span style="color: #5E62D1;">Como se ve a continuacin:</span></font>

```python
# En el siguiente ejemplo, se crea la variable llamada "mi_lista" seguido del operador igual (=) el cual  
# asigna a la variable lo que se encuentre a su derecha, en este caso una lista de nmeros enteros

mi_lista = [10,20,5,100,50]

# Despus de que se creo la lista y asigno a la variable "mi_lista" se utiliza la funcin print para mostrar 
# la lista de nmeros

print(mi_lista)
```

```python
type(mi_lista)
```

<font size="6"><span style="color: #33D6CC;">Indexacin</span></font>

<font size="3"><span style="color: #5E62D1;"> La indexacin se refiere a la forma en como se accede a los elementos de la lista u otros tipos de secuencias.</span></font>

<center><font size="3"><span style="color: #5E62D1;"><b>La indexacin se consigue a travs del ndice de los elementos</b> </span></font></center>



```python
mi_lista = ["Palabra", 50, 12.4] #Se crea la lista 
print(mi_lista[0]) #A travs de los corchetes se accede al elemento que se encuentra en el ndice 0
```

```python
print(mi_lista[2]) #A travs de los corchetes se accede al elemento que se encuentra en el ndice 2
```

```python
print(mi_lista[-2]) #A travs de los corchetes se accede al elemento que se encuentra en el ndice -2
```

<font size="6"><span style="color: #33D6CC;">Slicing</span></font>

<font size="3"><span style="color: #5E62D1;"> 
Anteriormente se accedio a un solo elemento a travs del ndice. Hacer **slicing** permite acceder a varios elementos a la vez.</span></font>


```python
mi_lista = [1,2,3,4,5,"a","b","c","d","e",-1,-2,-3,-4,-5]
mi_lista
```

```python
len(mi_lista)
```

```python
# Quiero obtener de mi_lista solo las letras
mi_lista[5:-5]
```

```python
# Quiero obtener de mi_lista los nmeros negativos
mi_lista[10:15]
```

```python
# Quiero obtener de mi_lista los nmeros negativos
mi_lista[10:]
```

```python
# Quiero obtener de mi_lista los nmeros negativos
mi_lista[-5:]
```

```python
# Quiero obtener de mi_lista los nmeros negativos
mi_lista[-5:15]
```

<font size="6"><span style="color: #33D6CC;">Mtodo **insert**</span></font>

<font size="3"><span style="color: #5E62D1;"> 
Se utiliza para insertar un elemento a la lista en la posicin que se desea a travs del ndice de dicha posicin.</span></font>


```python
mi_lista = ["Jorge","Braulio","Roberto", "Sandra","Jimena","Matilde", "Carmen"]
mi_lista.insert(3,"RODRIGO") # Insertar la cadena "RODRIGO" en el ndice  3
mi_lista
```

<font size="6"><span style="color: #33D6CC;">Mtodo **extend**</span></font>

<font size="3"><span style="color: #5E62D1;"> 
Se utiliza para agregar un conjunto de elementos a una lista.</span></font>


```python
mi_lista = ["Jorge","Braulio","Roberto", "Sandra","Jimena","Matilde", "Carmen"]
mi_lista.extend(["Rodrigo","Julio","Lorenzo"]) # Agregar los elementos "Rodrigo", "Julio" y "Lorenzo" a la lista
mi_lista
```

<font size="6"><span style="color: #33D6CC;">Mtodo **append**</span></font>

<font size="3"><span style="color: #5E62D1;"> 
Se utiliza para agregar un elemento al final de la lista.</span></font>


```python
mi_lista = ["Jorge","Braulio","Roberto", "Sandra","Jimena","Matilde", "Carmen"]
mi_lista.append("Julio") # Agregar el elemento "Julio" a la lista
mi_lista
```

<font size="6"><span style="color: #33D6CC;">Eliminar elementos de una lista</span></font>

```python
# Utilizando remove

mi_lista = ["Jorge","Braulio","Roberto", "Sandra","Jimena","Matilde", "Carmen"]
mi_lista.remove("Sandra")
mi_lista
```

```python
# Utilizando pop

mi_lista = ["Jorge","Braulio","Roberto", "Sandra","Jimena","Matilde", "Carmen"]
mi_lista.pop(3)
mi_lista
```

```python
# Utilizando del

mi_lista = ["Jorge","Braulio","Roberto", "Sandra","Jimena","Matilde", "Carmen"]
del mi_lista[3]
mi_lista
```

<font size="6"><span style="color: #33D6CC;">Mtodo **sort()**</span></font>

<font size="3"><span style="color: #5E62D1;"> 
Se utiliza para ordenar una lista.</span></font>


```python
mi_lista = ["Jorge","Braulio","Roberto", "Sandra","Jimena","Matilde", "Carmen"]
mi_lista.sort()
mi_lista
```

```python
mi_lista = ["Jorge","Braulio","Roberto", "Sandra","Jimena","Matilde", "Carmen"]
mi_lista.sort(reverse=True)
mi_lista
```

<font size="6"><span style="color: #33D6CC;">Mtodo **reverse()**</span></font>
