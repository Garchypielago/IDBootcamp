### Cread una función para automatizar el apartado anterior, esta función recive una url

```python
def receta_aut(enlace):

    url = enlace
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")


    Titulo = []
    Descripcion = []
    Comensales = []
    Tiempo = []
    Tipo = []
    Nivel = []
    IngredientesAux = []
    Ingredientes = []
    PasosAux = []
    Pasos = []

    rece = soup.find("main" )
    # Titulo
    try:
        Titulo.append(rece.find("h1").text.strip())
    except AttributeError:
        Titulo.append(np.nan)
    
    # Descripcion
    try:
        Descripcion.append(rece.find("div", attrs={"class":"intro"}).text.strip())
    except AttributeError:
        Descripcion.append(np.nan)
    
    # Comensales
    try:
        Comensales.append(rece.find("span", attrs={"class":"property comensales"}).text.strip())
    except AttributeError:
        Comensales.append(np.nan)
    
    # Tiempo
    try:
        Tiempo.append(rece.find("span", attrs={"class":"property duracion"}).text.strip())
    except AttributeError:
        Tiempo.append(np.nan)
    
    # Tipo
    try:
        Tipo.append(rece.find("span", attrs={"class":"property para"}).text.strip())
    except AttributeError:
        Tipo.append(np.nan)
    
    # Nivel
    try:
        Nivel.append(rece.find("span", attrs={"class":"property dificultad"}).text.strip())
    except AttributeError:
        Nivel.append(np.nan)

    # Ingredientes
    for i in rece.find_all("li", attrs={"class":"ingrediente"}):
        try:
            IngredientesAux.append(i.find("label").text.strip())
        except AttributeError:
            IngredientesAux.append(np.nan)
    Ingredientes.append(IngredientesAux)

    # Pasos
    for i in rece.find_all("div", attrs={"class":"apartado"}):   
        if i.find("div", attrs={"class":"orden"}) is not None:
            try:
                Paso = i.find("div", attrs={"class":"orden"}).text.strip() + ") " + i.find("p").text.strip()
                PasosAux.append(Paso)
            except AttributeError:
                PasosAux.append(np.nan)
    Pasos.append(PasosAux)
    N_Pasos = len(PasosAux)

    lista_completa = [Titulo, Descripcion, Comensales, Tiempo, Tipo, Nivel, Ingredientes, Pasos, N_Pasos]

    return lista_completa
```

```python
def recetas_df(num_r=100):
    lista_links_Total = []
    listas_completas = []

    for n in range(1,35):
        urlTotal = f"https://www.recetasgratis.net/Recetas-de-Ensaladas-listado_receta-4_{n}.html"
        responseTotal = requests.get(urlTotal)
        soupTotal = BeautifulSoup(responseTotal.text, "html.parser")

        for rece in soupTotal.find_all("div", attrs={"class":"resultado link"} ):
            lista_links_Total.append(rece.a["href"])
            if (len(lista_links_Total)>=num_r):
                break
        if (len(lista_links_Total)>=num_r):
            break

    for enlace in lista_links_Total:
        listas_completas.append(receta_aut(enlace))

    # Convertir listas_completas en un DataFrame
    data = {
        'Titulo': [lista[0][0] for lista in listas_completas],
        'Descripcion': [lista[1][0] for lista in listas_completas],
        'Comensales': [lista[2][0] for lista in listas_completas],
        'Tiempo': [lista[3][0] for lista in listas_completas],
        'Tipo': [lista[4][0] for lista in listas_completas],
        'Nivel': [lista[5][0] for lista in listas_completas],
        'Ingredientes': [lista[6][0] for lista in listas_completas],
        'Pasos': [lista[7][0] for lista in listas_completas],
        'N_Pasos': [lista[8] for lista in listas_completas]
    }
    
    df = pd.DataFrame(data)
    return df
```
