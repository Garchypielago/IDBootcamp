```python
url = "https://www.filmaffinity.com/es/topcat.php?id=new_th_es"

response = requests.get(url)
bool(response)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
titulos = []
duracion = []
posicion = []
categoria  = []
categorias  = []
director = []
casting = []
puntuacion = []
numero_votaciones = []

# , "class":"top-movie  first"
for peli in soup.find_all("div", attrs={"class":"top-movie"} ):

    duracion.append(peli.find("div", attrs={"class":"duration"}).text.strip())
    posicion.append(peli.find("div", attrs={"class":"pos"}).text.strip())
    titulos.append(peli.find("h3").text.strip())

    # # solucion metinedo la lista de categorias
    # if peli.find("span", attrs={"class":"type"}) is not None:
    #     for peli2 in peli.find_all("span", attrs={"class":"type"}):
    #         categorias.append(peli2.text.strip())
    # else:
    #     categoria.append(peli.find("a", attrs={"class":"genre"}).text.strip())

    # solucion metinedo la ult categoria
    if peli.find("span", attrs={"class":"type"}) is not None:
        categoria.append(peli.find_all("span", attrs={"class":"type"})[-1].text.strip())
    else:
        categoria.append(peli.find("a", attrs={"class":"genre"}).text.strip())

    director.append(peli.find("div", attrs={"class":"director"}).text.strip())
    casting.append(peli.find("div", attrs={"class":"cast"}).text.strip())
    puntuacion.append(peli.find("div", attrs={"class":"avg-rating"}).text.strip())
    numero_votaciones.append(peli.find("div", attrs={"class":"rat-count"}).text.strip())

print(
len(titulos),
len(duracion),
len(posicion),
len(categoria),
len(director),
len(casting),
len(puntuacion),
len(numero_votaciones))
```

```python
data = {
    'Orden': posicion,
    'Titulo': titulos,
    'Categoria': categoria,
    'Director': director,
    'Casting': casting,
    'Duración': duracion,
    'Puntuación': puntuacion,
    'Votaciones': numero_votaciones
}

df = pd.DataFrame(data)
df.head(5)
```
