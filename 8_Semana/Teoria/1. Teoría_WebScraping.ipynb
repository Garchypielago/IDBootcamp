# find_all: busca todos los tags del html
### ATENCIÓN: find_all NO devuelve un elemento tag, devuelve una LISTA de elementos tag

```python
lista_a = tag_login.find_all("a")
```

```python
lista_a[0]['href']
```

```python
for tag in lista_a: # recorro la lista de tags
    print(tag['href'])
```

```python
tags_a = soup.find_all("a") #Encontrar todos los elementos que sean hipervinculos
```

```python
len(tags_a)
```

```python
tags_a[113]['href'] # ahora podemos iterar por elementos y tratarlos por separado
```

```python
soup.find_all("a")[354]["href"]
# find busca el primero (me devuelve un elemento tipo soup)
# find_all busca todos (me devuelve un elemento tipo lista (cada posicion tiene un elemento soup))
```

```python
lista_links = []
for i in soup.find_all("a"): # esta función me devuelve una lista
    # i es un tag
    lista_links.append(i["href"])
    
#lista_links
```

```python
lista_links
```

```python
# Puedo hacerlo usando list comprehension para obtener una lista con solo el link del hipervinculo
listalinks = [i["href"] for i in soup.find_all("a")]
# listalinks
```

```python
titulos = [i.text for i in soup.find_all("a")]
```

```python
titulos
```

### Ejercicio: Encontrar los titulares y las noticias y sus hipervinculos de elpais.com

```python
url = "https://elpais.com"
response = requests.get(url)
print(response)
print(bool(response))
soup = BeautifulSoup(response.text, "html.parser")
```

```python
h2 = soup.find_all("h2")
len(h2)
```

```python
lista_ej2 = []
listalinks_ej2 = []
listatitulares_ej2 = []
for h2 in soup.find_all("h2"):
    # Verificar si el elemento <span> existe dentro del <h2>
    if h2.span is not None:
        lista_ej2.append(h2)
        listalinks_ej2.append(h2.a["href"])
        listatitulares_ej2.append(h2.a.text)
```
