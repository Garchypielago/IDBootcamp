```python
from operagxdriver import start_opera_driver
import os

# Ruta a los archivos de ejecutable de Opera y OperaDriver
opera_browser_exe = r"C:\Users\garch\AppData\Local\Programs\Opera GX\launcher.exe"
opera_driver_exe = r"operadriver.exe"

# Ruta del directorio de usuario
userdir = r"C:\Users\garch\opera_profile"

# Crear el directorio si no existe
os.makedirs(userdir, exist_ok=True)

# Verificar si los archivos existen y se tienen permisos
assert os.path.isfile(opera_browser_exe), f"Opera executable not found at {opera_browser_exe}"
assert os.access(opera_browser_exe, os.R_OK), f"Permission denied for Opera executable at {opera_browser_exe}"
assert os.path.isfile(opera_driver_exe), f"OperaDriver executable not found at {opera_driver_exe}"
assert os.access(opera_driver_exe, os.R_OK), f"Permission denied for OperaDriver executable at {opera_driver_exe}"

# Iniciar el driver de Opera
driver = start_opera_driver(
    opera_browser_exe=opera_browser_exe,
    opera_driver_exe=opera_driver_exe,
    userdir=userdir,
    arguments=(
        "--no-sandbox",
        "--test-type",
        "--no-default-browser-check",
        "--no-first-run",
        "--disable-dev-shm-usage",
        "--disable-extensions",
        "--remote-debugging-port=9222",
        "--start-maximized",
    )
)
```
