```python
import shutil
import os
```

```python
imagen = open('imagen_descargada3.jpg', 'wb') # abre una imagen en la misma carpeta, si ya existe la sobreescribe, wb es de write binary 
resp = requests.get(url, stream=True) 
# El parámetro stream=True indica que la respuesta no se descarga de inmediato, bueno para imagenes/archivos grandes
resp.raw.decode_content = True
# el atributo raw es porque es sin procesar
# decode_content = True indica que se deben descodificados segun el origen de la request
shutil.copyfileobj(resp.raw, imagen)
imagen.close()
```

### Podemos cambiar el Header
Los HTTP headers son la parte central de los HTTP requests y responses, y transmiten información acerca del navegador del cliente, de la página solicitada, del servidor, etc.

```python
HEADER = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
```

```python
resp = requests.get(url, headers=HEADER)
print(resp)
# print(resp.text) # esto lo convertiriamos en un soup con BeautifulSoup
```

```python
# Hay otras librerías que también pueden funcionar para hacer request, y en algunos casos se recomiendan
```

https://stackoverflow.com/questions/2018026/what-are-the-differences-between-the-urllib-urllib2-urllib3-and-requests-modul

```python
import urllib.request as urllib2
```

```python
def get_page(url):
    req = urllib2.Request(url, headers=HEADER)
    page = urllib2.urlopen(req, timeout=200).read()
    soup = BeautifulSoup(page, "html.parser")
    return soup

soup = get_page(url)
```
