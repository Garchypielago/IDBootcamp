<font size="7"><span style="color: #33D6CC;">&iquest;Qu&eacute; es el bucle WHILE?</span></font>

<font size="3"><span style="color: #5E62D1;"> Es una función de programación que nos permite iterar un bloque de código, similar a la función **for**. </span></font>

<font size="3"><span style="color: #5E62D1;">La gran diferencia es que el bucle **for** es posible conocer el número de iteraciones mientras que al utilizar el bucle while no sabemos cuantas iteraciones se van a realizar.</span></font>

<center><font size="3"><span style="color: #5E62D1;"><b>¿Cuándo utilizar un bucle WHILE?</b></span></font></center>

<center><font size="3"><span style="color: #5E62D1;">Se utiliza cuando se quiere que un bloque de código se ejecute hasta que cierta condición se cumpla.</span></font></center>
<p>&nbsp;</p>
<font size="3"><span style="color: #5E62D1;"><b>El flujo de su funcionamiento es el siguiente:</b></span></font>


<ol>
<li><font size="3"><span style="color: #5E62D1;">Determinar si la condición es cierta o falsa.</span></font></li>
<li><font size="3"><span style="color: #5E62D1;">Si la condición es falsa, no ejecuta el bucle y continúa la secuencia del código.
</span></font></li>
<li><font size="3"><span style="color: #5E62D1;">Por lo contrario, si la condición es verdadera, se ejecuta el bloque de código dentro del bucle y se regresa al paso 1.
</span></font></li>
</ol>

<font size="6"><span style="color: #33D6CC;">Sintaxis del bucle while</span></font>


```python
edad = 50

while edad<18:
    print("Hola mundo")

print(2-1)
```

```python
## Ejemplo 1 
# Pedir la contraseña a un usuario

contraseña = "No lo sé"

while contraseña != "admin1234":
    contraseña = input("Ingresa la contraseña: ")
    if contraseña != "admin1234":
        print("Tu contraseña es incorrecta")

print("Tu contraseña es correcta")
```

```python
## Ejemplo 2
# While como FOR
# Recorrer la siguiente lista
lista = ["a", "b", "c", "d", "e"]
x = 0

while x != len(lista):
    print(lista[x])
    x+=1
```

```python
for i in lista:
    print(i)
```

```python
## Ejemplo 3
# While como FOR utilizando el booleano True
# Recorrer la siguiente lista
lista = ["a", "b", "c", "d", "e"]
x = 0

while True:
    print(lista[x])
    x+=1
    print(x)
    if x == len(lista):
        break
```

```python
# Ejemplo 4
# While en una sola linea. 
# Se utiliza para iterar n cantidad de veces una línea de código 

numero = 0
while numero < 7: numero += 1; print("El número es: " + str(numero)); 
```
