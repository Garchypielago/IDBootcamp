# Column transformer (aplicar diferentes pipelines a diferentes columnas)

```python
X = df.drop(columns=["SalePrice"])
Y = df.SalePrice
```

```python
numeric_features = X._get_numeric_data().columns
```

```python
numeric_features
```

```python
categorical_features = X.drop(columns=numeric_features).columns
```

```python
X.drop(columns=numeric_features).columns
```

#### Pipeline para numericas

```python
# paso de transformar columnas numericas
numeric_transformer = Pipeline(
    steps = [("imputer", SimpleImputer(strategy="median")), 
           ("scaler", StandardScaler())]
)
# scaler=StandardScaler()
# numeric_transformer = make_pipeline(SimpleImputer(strategy="median"), scaler)
```

```python
numeric_transformer
```

#### Pipeline para categoricas

```python
# paso de transformar columnas categoricas
categorical_transformer = Pipeline(
    steps=[("imputer_cat", SimpleImputer(strategy="most_frequent")),
           ("oneHotEncoder", OneHotEncoder(drop = "first", handle_unknown="ignore"))]
)

# categorical_transformer= make_pipeline(SimpleImputer(strategy="most_frequent"), 
#                                        OneHotEncoder(drop = "first", handle_unknown="ignore"))
categorical_transformer
```

### Juntamos los dos pipelines en paralelo

```python

# Paso de transformar las columnas
preprocessor = ColumnTransformer(
    transformers=[
        ("pipeline para numericas", numeric_transformer, numeric_features),
        ("pipeline para categoricas", categorical_transformer, categorical_features)
        # 3 elementos (nombre, pipeline, lista_de_nombres_columnas)
    ]
)
preprocessor
```

#### Finalmente juntamos el ColumnTransformer a un modelo con un pipeline, poniendo en un objeto todo el proceso completo

```python
# Pipeline de las diferentes columnas
proceso_completo = Pipeline(
    steps = [("preprocessor", preprocessor), 
           ("linearModel", LinearRegression())]
)
```

```python
# class mi_preprocesamiento():
#     def __init__
    
#     def fit()
    
#     def transform()
```

```python
proceso_completo
```

#### Una vez definido todo el proceso, es cuando entrenamos todo

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# aqui se ajusta todo
proceso_completo.fit(X_train, y_train)

y_hat = proceso_completo.predict(X_test)
y_hat_train = proceso_completo.predict(X_train)
```

```python
print("model error:", mean_squared_error(y_test,y_hat, squared = False))
```

```python
print("model r2_score:", r2_score(y_test,y_hat))
```

```python
print("model r2_score:", r2_score(y_train,y_hat_train))
```
