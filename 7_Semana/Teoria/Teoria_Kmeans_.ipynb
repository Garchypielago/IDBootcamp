```python
##### K Means
import matplotlib.pyplot as plt
import seaborn as sns; 
import numpy as np
import pandas as pd
```

```python
#Genero datos esfericos
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300,
                       centers=3,
                       cluster_std=0.7)

plt.scatter(X[:, 0], X[:, 1], s=50)
```

```python
y_true
```

```python
X.shape
```

```python
#Estimo Kmeans
from sklearn.cluster import KMeans
#inicializo el objeto
kmeans = KMeans(n_clusters=4) # 

kmeans.fit(X) # ya lo hace todo
# 1: Genera centroides aleatorios
# 2: Calcula distancias y asigna cada fila a un centro
# 3: recalcula los centroides a artir de los labels asinados
# repite los pasos 2-3
centers = kmeans.cluster_centers_ # estos son los centros
```

```python
centers
```

```python
kmeans.predict([[-2,-3],
                [0, 0],
                [10,10],
                [10,0]])
```

```python
# Aqui coloca las muestras en sus grupos
y_kmeans = kmeans.predict(X) # yo coloco una muestra
y_kmeans
```

```python
centers
```

```python

# Pinto los puntos (las filas)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Pinto los CENTROS
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
print("Inercia del clustering", kmeans.inertia_)
# print(centers)
```

```python
#Estimo Kmeans
from sklearn.cluster import KMeans
#inicializo el objeto
kmeans = KMeans(n_clusters=10)
kmeans.fit(X) # ya lo hace todo
centers = kmeans.cluster_centers_ # estos son los centros
centers
```

```python
# Aqui coloca las muestras en sus grupos
y_kmeans = kmeans.predict(X) # yo coloco una muestra
# Pinto los puntos (las filas)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
# Pinto los CENTROS
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
print("Inercia del clustering", kmeans.inertia_)
print(centers)
```

```python
sse = [] # lista está vacia y voy a llenarla de las inercias o sse de cada separacion
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
```

```python
sse
```

```python
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

# Silhouette score

```python
from sklearn.metrics import silhouette_score
# silhouette va des de -1 a 1
# nos dice como de similar es una muestra a su cluster
silhouette_scores = []

# Tiene en cuenta 2 factores
# - Como de cerca estan los puntos de un grupo
# - Como de lejos estan de los otros grupos

# Empezamos con 2
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    labels=kmeans.predict(X)
#     score = silhouette_score(X, kmeans.labels_)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)
```

```python
plt.plot(range(2, 11), silhouette_scores)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
from sklearn import datasets
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Create scaler: scaler
scaler = StandardScaler()
# scaler.fit(X)
# X = scaler.transform(X)
```

```python
# Create KMeans instance: kmeans
kmeans = KMeans(n_clusters = 3)

# Create pipeline: pipeline
pipeline = make_pipeline(scaler, kmeans)

# Fit the pipeline to samples
pipeline.fit(X)
```

```python
pipeline.predict(X)
```

```python
X=pd.DataFrame(X)
```

```python
X['label']=pipeline.predict(X)
```
