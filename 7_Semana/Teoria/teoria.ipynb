```python
##### K Means
import matplotlib.pyplot as plt
import seaborn as sns; 
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("BostonHousing.csv")
```

```python
df.chas.value_counts()
```

```python
X=df.drop(columns=['ID','chas'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
X_esc = scaler.transform(X)
```

```python
X_esc
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters = 3)
```

```python
kmeans.fit(X_esc)
```

```python
from sklearn.metrics import silhouette_score
labels=kmeans.predict(X_esc)
silhouette_score(X_esc, labels)
kmeans.inertia_
```

```python
sse=[]
silhouettes=[]
for k in range(2, 10):
    kmeans=KMeans(n_clusters = k)
    kmeans.fit(X_esc)
    
    labels=kmeans.predict(X_esc)
    silhouettes.append(silhouette_score(X_esc, labels))
    sse.append(kmeans.inertia_)
```

```python
sse
silhouettes
plt.plot(range(2, 10),sse)
```

```python
plt.plot(range(2, 10),silhouettes)
```

```python
kmeans=KMeans(n_clusters = 2)
kmeans.fit(X_esc)

labels=kmeans.predict(X_esc)
labels
```

```python
X['cluster']=labels
```

```python
X.groupby("cluster").mean()
```

```python
sns.relplot(data=X, x="medv", y =  "crim", hue='cluster')
```
