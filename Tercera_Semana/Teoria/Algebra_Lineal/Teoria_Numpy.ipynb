```python
MM[0,:,:]
# de el conjunto 0
# sacame todas las filas
# para todas las columnas
```

```python
MM[1,:,:]
# de el conjunto 1
# sacame todas las filas
# para todas las columnas
```

```python
MM[2,:,:]
# de el conjunto 2
# sacame todas las filas
# para todas las columnas
```

```python
MM.min(axis=2)
# saca el min de cada posicion de cada conjunto
```
