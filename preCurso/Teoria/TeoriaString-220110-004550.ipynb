<font size="7"><span style="color: #33D6CC;">&iquest;Qu&eacute; es un string?</span></font>

<font size="3"><span style="color: #5E62D1;">Si quieres usar cualquier tipo de texto en Python, usarás un <strong>(string)</strong>. Es un tipo básico de python, se crea <strong>introduciendo texto entre comillas simples o dobles.</strong> </span></font>

<font size="3"><span style="color: #5E62D1;">El uso de comillas simples <strong> ' ' </strong> o comillas dobles <strong> " "</strong> no afecta al funcionamiento del propio <strong> string </strong> </span></font>

<font size="3"><span style="color: #5E62D1;">Ahora, haciendo uso de la función <strong> print </strong> imprimiremos la cadena <strong> Hola Mundo! </strong> (un clásico en programación)</span></font>

```python
print("Hola Mundo!")
#También podemos utilizar comillas simples
print('Hola Mundo!')
```

<font size="3"><span style="color: #5E62D1;"> Algunos caracteres no pueden inlcuirse directamente dentro de un <strong> string </strong>, por ejemplo, las comillas que estamos usando para definir dicho <strong> string </strong> porque marcarían el final del propio <strong> string </strong> </span></font>

<font size="3"><span style="color: #5E62D1;"> Para solucionarlo utilizaremos <strong> \ (backslash) </strong> antes del caracter. </span></font>

<font size="3"><span style="color: #5E62D1;"> Si estamos usando comillas dobles, las comillas simples no necesitan el backslash y viceversa.</span></font>

<font size="3"><span style="color: #5E62D1;"> Podemos usar <strong> backslash </strong> para introducir ciertos caracteres especiales como tabulaciones, retornos de carro, caracteres Unicode...</span></font>

```python
"hola a todos, pero en especial a \"ti\" "
```

```python
print('Sócrates dijo: "El conocimiento empieza en el asombro"')
```

```python
# El caracter especial \n nos servirá para crear un salto de linea:
print('Aqui\ntermina\nla\nprimera\nlinea\ny\naqui\nempieza\nla\nsegunda')
```

```python
# De manera parecida podemos utilizar \t para generar una tabulación.
print('Primera sección \t sección después de la tabulación')
```

<font size="3"><span style="color: #5E62D1;"> Las cadenas de caracteres en Python <strong> tienen índices </strong> con los que vamos a poder crear <strong> subcadenas. </strong></span></font>

<font size="3"><span style="color: #5E62D1;"> Mediante el uso de corchetes, podemos acceder a los elementos de una cadena, por ejemplo, obtener la "M".</span></font>

```python
cadena = 'Hola Mundo'
print(cadena[-5]) #NOTA: El resultado vuelve a ser un string
```

```python
print(cadena[-10]) #Nos devuelve el último elemento
```

<font size="3"><span style="color: #5E62D1;"> Llamaremos <strong> slicing </strong> de cadenas al método mediante el cual obtenemos una subcadena. Lo aplicaremos con corchetes</span></font>

```python
print(cadena[3:6]) #NOTA: Podemos obviar el 0 y tomará el inicio de la cadena
```

```python
print(cadena[:])
```

```python
print(cadena[-2:]) #Esto nos devuelve el string sin las últimas 2 posiciones
```

<font size="3"><span style="color: #5E62D1;"> Llamaremos <strong> stride </strong> al <strong> paso </strong> con el que vamos a recorrer nuestra cadena. En los ejemplos anteriores, no se especificaba y, por tanto, se consideraba como 1, es decir, recorríamos los caracteres de la cadena de uno en uno. <strong> Veamos como podemos devolver toda la cadena de dos en dos caracteres. </strong> </span></font>

<font size="3"><span style="color: #5E62D1;"> <strong>cadena[inicio:fin:paso]</strong> </span></font>

```python
print(cadena[0:10:1]) #Podemos utilizar un paso diferente
```

```python
print(cadena[5:8:2])
```

<font size="3"><span style="color: #5E62D1;"> Veamos algunos <strong> métodos </strong> útiles para utilizar en las cadenas de caracteres: </span></font>

<ul>
<font size="3"><li><span style="color: #5E62D1;"><strong> .find() </strong>: sirve para encontrar un caracter o un conjunto de caracteres dentro de la cadena.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .startswith() y .endswith() </strong>: comprueba si la cadena empieza o termina con un caracter o conjunto de caracteres. Devuelve un valor booleano.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .split() </strong>: divide la cadena y devuelve una lista con cada una de las partes de la cadena.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .upper() y .lower() </strong>: convierte la cadena a mayúsculas o minúsculas respectivamente.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .replace() </strong>: reemplaza una parte de la cadena por otra dada.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .strip() </strong>: elimina los caracteres del inicio o fin de la cadena.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .count() </strong>: cuenta el número de apariciones de un caracter o un conjunto de caracteres en la cadena.</span></li></font>
</ul>

```python
cadena = 'Hello world!'
cadena.find("w") #devuelve el índice de la letra "w"
```

```python
cadena = 'Hello world!'
print(cadena.startswith('H')) #devuelve True, puesto que la cadena empieza con H.
print(cadena.endswith('d')) #devuelve False, puesto que la cadena termina por !
```

```python
cadena = 'Hello world!Hola'
print(cadena.split(' ')) #divide la cadena por espacios y crea una lista
```

```python
cadena = 'Hello world!'
print(cadena.upper()) #imprime la cadena en mayúsculas
print(cadena.lower()) #imprime la cadena en minúsculas
```

```python
cadena = 'Hello world!'
print(cadena.replace('world', 'mundo')) #imprime la cadena cambiando Hello por Hola
```

```python
#La función strip es muy útil para eliminar espacios al inicio o final de nuestras cadenas de caracteres
nueva_cadena = '       Hola       '
print(nueva_cadena)
```

```python
print(nueva_cadena.strip(' '))
```

```python
cadena = 'Hello world!'
print(cadena.count('l')) #cuenta el número de "l" --> 3
```

```python
cadena = 'Esta cadena tiene 31 caracteres'
len(cadena)
```
