<font size="6"><span style="color: #33D6CC;">**Sintaxis del condicional if - elif ... -elif - else.**</span></font>

<font size="3"><span style="color: #5E62D1;">Mltiples condiciones.</span></font>


```python
# Ejemplo 3
# Se imprime si un nmero es negativo, cero o positivo. 

numero = 50

if numero < 0:
    print("nmero negativo")
elif numero == 0:
    print("nmero igual a cero")
else:
    print("nmero positivo")
```

<font size="6"><span style="color: #33D6CC;">Condicionales anidados</span></font>

<font size="3"><span style="color: #5E62D1;">Un condicional anidado significa que dentro de un condicional podemos tenemos otro condicional.</span></font>


```python
# Ejemplo 4
# Se imprime la raza de un perro de acuerdo con las caractersticas del perro

altura_perro = 25 #cm
peso = 20 #kg

if altura_perro>50:
    if peso > 20:
        print("Perro de raza grande")
    elif 10<=peso<=20:
        print("Perro de raza mediana")
    else:
        print("Perro similar a un galgo")
else:
    if peso<10:
        print("Perro de raza mini")
    else:
        print("Perro de raza mediana o pequea")
```

```python
# Ejemplo 5
# Se imprime la raza de un perro de acuerdo con las caractersticas del perro

altura_perro = 25 #cm
peso = 20 #kg

if altura_perro>50 and peso > 20:
        print("Perro de raza grande")
elif altura_perro>50 and 10<=peso<=20:
        print("Perro de raza mediana")
elif altura_perro>50 and peso<10:
        print("Perro similar a un galgo")
elif altura_perro<=50 and peso<10:
        print("Perro de raza mini")
elif altura_perro<=50 and peso>=10:
        print("Perro de raza mediana o pequea")
```

<font size="6"><span style="color: #33D6CC;">Condicional en una lnea</span></font>

<font size="3"><span style="color: #5E62D1;">Se suele utilizar para asignar el valor a una varible dada una condicin.</span></font>


```python
# Ejemplo 6
# Si la viarble estacin es igual a "invierno", a la variable vestuario se le asigna la cadena "abrigo"
# Si la variable estacin es igual a "verano", a la variable vestuario se le asigna la cadena "camiseta"
# Utilizando un if - else.

estacion = "invierno"
if estacion == "invierno":
    vestuario = "abrigo"
else:
    vestuario = "camiseta"
print(vestuario)
```

```python
#Utilizando el condicional if - else en una sola lnea

estacion = "verano"
vestuario = "abrigo" if estacion=="invierno" else "camiseta"
vestuario
```
