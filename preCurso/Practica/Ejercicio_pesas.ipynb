# Ejercicio de lógica:
Imaginemos que disponemos de 8 bolas iguales a la vista y que tenemos una balanza con dos bandejas para poder pesarlas.

Supongamos que sabemos que una de las bolas pesa más que el resto pero no sabemos cuál es. ¿Cuál es el mínimo número de pesadas que necesitamos para poder saber con total seguridad cuál es esa bola? ¿Cuál sería el procedimiento a seguir?

Una vez que tengáis el procedimiento, debéis de pasar este algoritmo a python. Poned una lista de ejemplo hecha por vosotros con los pesos de las bolas (por ejemplo L=[1, 1, 1, 1, 1, 1.1, 1,1,], la bola está en la posición 5).

```python
L=[1, 1, 1, 1, 1, 1, 1, 2]
contador = 0
for i in range(int(len(L)/2-1)):
    L.sort(reverse=True)
    L1=L[int(len(L)/2):int(len(L))]
    if L[0] > L1[0]:
        L = L[0:int(len(L)/2)]
    else:
        L = L1
    contador += 1
contador
```
