<font size="3"><span style="color: #5E62D1;"> 7. Dada la siguiente tupla: (9,8,7,5,"a",3,6,5), encuentra la forma para eliminar la letra "a". </span></font>


```python
tupla = (9,8,7,5,"a",3,6,5)
tupla = list(tupla)
for t in tupla:
    if type(t) == type('a'):
        tupla.remove(t)

tupla
```
