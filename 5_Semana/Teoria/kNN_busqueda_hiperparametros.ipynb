#### Haz la función calcula_knn que divida el dataset en train y test y haga un modelo knn de k vecinos

```python
def calcula_knn(X, y, k, n=10, div=0.3):
    L_R2test=[]
    for i in range(n):
        X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=div)

        objeto_escalador = StandardScaler()
        objeto_escalador.fit(X_train)

        X_train[:] = objeto_escalador.transform(X_train)
        X_test[:] = objeto_escalador.transform(X_test)

        modelo = KNeighborsRegressor(k)
        modelo.fit(X_train, y_train)

        yhat = modelo.predict(X_test)
        L_R2test.append(r2_score(y_test, yhat))

    return L_R2test
```

```python
calcula_knn(X,y,30)
```
