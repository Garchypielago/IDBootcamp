```python
def maximo(L):
    maxi = L[0] # considero que el maximo es el primero
    for elem in L:
        if elem > maxi: # si el elem es mayor que el maximo, cambiamos el maximo
            maxi = elem
    return maxi
```

### Ej. Dada una lista, cread una funcion que encuentre la lista de elementos unicos

```python
L=[1,1,2,2,3,3]
L_unicos=[]
for elem in L:
    if elem not in L_unicos:
        L_unicos.append(elem)
L_unicos
```

```python
# L = [1,1,1,2,2,3,3] => [1,2,3]
# solo con for, "in" y append()
# no vale utilizar sets
# unicos([1,2,3,3,4,4,1]) -> [1,2,3,4]
# Que recive? una lista con numeros repetidos
# Que devuelve? una lista con numeros sin repetir
def unicos(L):
    L_unicos=[] 
    for elem in L:
        if elem not in L_unicos:
            L_unicos.append(elem)
    
    return L_unicos
```

### Ej. Haced una funcion que dado un numero devuelva True si es primo y False si no

```python
n=2
esprimo=True
if n == 1: # no es primo
    esprimo=False
for i in range(2,n): # vamos a ver si existe algun divisor de n
    if n%i==0: #8/2, si entro aqui, no es primo
        esprimo=False
        break
esprimo
```

```python
# Primo = que solo lo dividan 1 y el mismo.
# es_primo(7) -> True
# es_primo(1) -> False
# es_primo(4) -> False
# un numero n es primo si solo se puede dividir entre 1 y n, pero 1 no
def es_primo(n):
    esprimo=True
    if n <= 1: # no es primo
        esprimo=False
    for i in range(2,n): # vamos a ver si existe algun divisor de n
        if n%i==0: #8/2, si entro aqui, no es primo
            esprimo=False
            break
    return esprimo
```

```python
def es_primo(n):
    if n <= 1: # no es primo
        return False
    for i in range(2,n): # vamos a ver si existe algun divisor de n
        if n%i==0: #8/2, si entro aqui, no es primo
            return False # 
    # si llego al final, es primo!
    return True
```

## Argumentos por referencia
* Si nuestra función modifica internamente (como el append) el objeto que le hemos pasado como argumento, también se modifica fuera de la función
* si SOBREESCRIVIMOS ese objeto dentro de la función, ya no afectará al objeto de fuera

```python
x=5 # esto no esta dentro de la funcion
def func(x): # recive una x, que funciona localmente en la funcion
    x=10 # se sobreescrive
    return x
print(func(x)) #10
print(x) # sigue siendo 5
```

```python
# Pasar por referencia
def add(L, n):# esta funcion, añadae un numero a una lista
    L.append(n) # utilizar una función que modifica internamente
    # no hay return
    return L
lista=[1,2,3,4]
n=10
add(lista, n) # le paso la referencia de memoria
```
