# agregate
## agregar a un mismo grupby, varias funciones de agrupado

```python
f = lambda x: np.sum(x**2)
# apply --> a una columna, a cada valor por separdo
# con el agregate ("agg") puedo aplicar una funcion externa, pero considerando
# que tiene que ser una funcion de grupo 

df.groupby("Continente")[[2010,2011,2012]].agg(["min", "max",f,np.mean])
```

```python
df5=df.groupby("Continente").agg({2010:["min","max","sum"],2011:["min","max"],2012:["min"]})
df5
```

```python
df5 = df5.reset_index()
df5
```
