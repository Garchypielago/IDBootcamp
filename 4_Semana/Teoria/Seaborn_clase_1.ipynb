```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

# Pairplot

```python
sns.pairplot(iris) # genera modelo vacio
plt.show() # 
```

```python
sns.pairplot(iris, vars=["sepal_length", "sepal_width"]) 
plt.show() 
```

```python
sns.pairplot(iris, vars=["sepal_length", "sepal_width", "species"]) 
plt.show() 
# esta es pruba para ver diferencia con la de abajo
```

```python
sns.pairplot(iris, vars=["sepal_length", "sepal_width"], hue="species") 
plt.show() 
```

```python
sns.barplot(y="sepal_length", x="species", data=iris, estimator=np.mean)
plt.show()
```

```python
sns.barplot(y=iris.iloc[:,1], x=iris["species"])
plt.show()
```

```python
df1 = pd.read_excel("..\\Practica\\Airbnb_NYC.xlsx",sheet_name = "NYC1")
df2 = pd.read_excel("..\\Practica\\Airbnb_NYC.xlsx",sheet_name = "NYC2")
```

```python
df = pd.merge(left=df1, right=df2, on="id")
```

```python
df.columns
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x="room_type", y="price", hue="municipio", data=df, palette="pastel")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.barplot(x="municipio", y="price", hue="room_type", data=df, palette="pastel")
plt.show()
```

```python
fig, axes = plt.subplots(2,1,figsize=(15,8))
sns.barplot(ax=axes[0], x="municipio", y="price", hue="room_type", data=df, palette="pastel")
sns.barplot(ax=axes[1], x="price", y="room_type", hue="municipio", data=df, palette="pastel")
plt.show()
```

## Pandas tiene seaborn incorporado

```python
df.groupby("room_type").size().plot(kind="bar")
```

```python
plt.figure(figsize=(12,3))

plt.subplot(1,3,1) # 1 fila, 3colums, subgrafo 1(el primer grafic en la primera columna)
df.groupby("room_type")["price"].mean().plot.bar()

plt.subplot(1,3,2)# 1 fila, 3colums, subgrafo 2(el segundo grafic en la segunda columna)
df.groupby("room_type")["price"].mean().plot(kind="line")

plt.subplot(1,3,3)# 1 fila, 3colums, subgrafo 1(el tercer grafic en la tercera columna)
df.groupby("room_type")["price"].mean().plot()
```

# Countplot

```python
fig, ax = plt.subplots(figsize=(6,4))
sns.countplot(x="municipio" ,data=df, hue="room_type", palette="Blues_r")
plt.show()
```

```python
sns.displot(x=np.log(df.price), color="green")
plt.show()
```

```python
sns.displot(x=df.price, color="green")
plt.show()
```

# Relplots, Joinplots, Catplots, Regplots

## Relplots

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y ="tip", hue="smoker", style="time", size="size", sizes=(100,200), data=tips, height=6)
plt.show()
```

```python
sns.relplot(x="total_bill", y ="tip", hue="smoker", style="time", size="size", sizes=(100,200), data=tips, height=6, col="time", row="sex")
```

# Joinplot

```python
sns.jointplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
plt.show()
```

```python
sns.kdeplot(x="total_bill", y="tip", data=tips)
plt.show()
```

## Catplots

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.catplot(x="event", y="signal", data=fmri, hue="region")
plt.show()
```

```python
sns.catplot(x="event", y="signal", data=fmri, kind="box", hue="region")
plt.show()
```

```python
sns.catplot(x="event", y="signal", data=fmri, hue="region", kind="boxen")
plt.show()
```

```python
sns.catplot(x="event", y="signal", data=fmri, hue="region", kind="violin")
plt.show()
```

## Regplots

```python
sns.lmplot(x="timepoint", y="signal", data=fmri)
```
