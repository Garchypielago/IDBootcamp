```python
var = "LSepal"
var_titulo = "Longitud del sÃ©palo"
#creo subgrÃ¡ficos con seaborn
fig, ax = plt.subplots (1,3, figsize=(15,3))
#Boxplot
sns.boxplot(data=df[var], ax=ax[0], orient="h", width=0.4)
ax[0].set_xlabel(var_titulo)
ax[0].set_title("Boxplot de " + var_titulo)
#Histograma
sns.histplot(data=df[var], ax=ax[1])
ax[1].set_xlabel(var_titulo)
ax[1].set_ylabel("Frecuencia")
ax[1].set_title("Histograma de " + var_titulo)

#Curva de Densidad
sns.kdeplot(data=df[var], ax=ax[2], fill=True)
ax[2].set_xlabel(var_titulo)
ax[2].set_ylabel("Densidad")
ax[2].set_title("Densidad de " + var_titulo)
plt.show()
```

## 1.Longitud Sepalo

```python
var = "LSepal"
var_titulo = "Longitud del sÃ©palo"
#creo subgrÃ¡ficos con seaborn
fig, ax = plt.subplots (1,3, figsize=(15,3))

# Histograma
sns.histplot(data=df_setosa, x=var, label="Setosa", ax=ax[0])
sns.histplot(data=df_versicolor, x=var, label="Versicolor", ax=ax[0])
sns.histplot(data=df_virginica, x=var, label="Virginica", ax=ax[0])
ax[0].set_xlabel(var_titulo)
ax[0].set_ylabel("Frecuencia")
ax[0].legend()

# Densidad
sns.kdeplot(data=df_setosa, x=var, label="Setosa", ax=ax[1])
sns.kdeplot(data=df_versicolor, x=var, label="Versicolor", ax=ax[1])
sns.kdeplot(data=df_virginica, x=var, label="Virginica", ax=ax[1])
ax[1].set_xlabel(var_titulo)
ax[1].set_ylabel("Densidad")
ax[1].legend()

# # Boxplot
# sns.boxplot(data=df_setosa, x=var, label="Setosa", ax=ax[2])
# sns.boxplot(data=df_versicolor, x=var, label="Versicolor", ax=ax[2])
# sns.boxplot(data=df_virginica, x=var, label="Virginica", ax=ax[2])
# ax[2].set_xlabel(var_titulo)
# ax[2].set_title(var_titulo)

plt.show()
```
