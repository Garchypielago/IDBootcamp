```python
# Utilización de un bucle for para mostrar 6 plots
x = np.linspace(-10, 10, 100)
y = x
for index, col in enumerate(["green","blue","red","black","brown","magenta"]):
    # (i, item)
    # en cada iteración, un grafico
    # print(index,col)
    plt.figure(figsize=(3,3))
    plt.plot(x, y**(index+1), color=col)  # x, y(=x**i, i = 1, (x,x), (x,x**2))
    plt.show()
```

```python
# subplot es una función que se utiliza para hacer varios plot 
plt.figure(figsize=(15,10))
x = np.linspace(-5, 5, 100)
y = x

# Como parametros de entrada para subplot vamos a tener 3 diferentes:
# num filas, num de col, el num de la grafica

plt.subplot(2,3,1) # abro un subplot para ir metiendo cosas de este plot
plt.plot(x, y**1, color="green") 
# abro otro
plt.subplot(2,3,2)  
plt.plot(x, y**2, color="blue")  
plt.subplot(2,3,3)  
plt.plot(x, y**3, color="red")  
plt.subplot(2,3,4)
plt.plot(x, y**4, color="magenta")  
plt.subplot(2,3,5)  
plt.plot(x, y**5, color="black")  
plt.subplot(4,3,9)  
plt.plot(x, y**6, color="brown") 
plt.subplot(4,3,12)  
plt.plot(x, y**6, color="brown") 
plt.show()  
```

```python
# subplot es una función que se utiliza para hacer varios plot 
plt.figure(figsize=(10,10))
x = np.linspace(-5, 5, 100)
y = x

# Como parametros de entrada para subplot vamos a tener 3 diferentes:
# num filas, num de col, el num de la grafica

plt.subplot(2,2,1) # abro un subplot para ir metiendo cosas de este plot
plt.plot(x, y**1, color="green") 
# abro otro 
plt.subplot(2,2,3)  
plt.plot(x, y**3, color="red")  
plt.subplot(1,2,2)
plt.plot(x, y**4, color="magenta")   
plt.subplot(4,2,6)
plt.plot(x, y**4, color="magenta")  
plt.show()  
```

```python
x = np.linspace(-5, 5, 20)
y = x

for k, col in enumerate(["green","blue","red","black","brown","magenta"]):
    plt.subplot(2, 3, k+1)
    plt.plot(x, y**(k+1), color=col)  
plt.show()
```

```python
# Creamos la figura "fig" donde se le van agregar 3 subplots utilizando la variable "ax"
# subplot != subplots
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20,6))
# plt.figure()
# fig tiene las caracteristicas generales de la grafica "grande"
x = np.linspace(0, 10, 100)
ax[0].plot(x, np.sin(x))        
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.tan(x))
plt.show()
```

```python
#Utilizamos un bucle for para llenar cada uno de los subplots de la figura 

x = np.linspace(-10, 10, 100)
y = x 

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(20,16))
k=1 # exponente
for row in ax: # recorro la lista ax( son 2 listas) 
    for ax1 in row: # recorro cada lista ( 3 subplots)
        ax1.plot(x, y**k, 'blue')
        ax1.set_title("hola"*k) # la función no es la misma en 1 grafico general que en axis de subplots
        ax1.tick_params(labelsize=15, labelrotation = 45)
        k+=1

plt.show()
```

```python
#Subplot en una figura de 2 x 2 
k=1
x = np.linspace(-10, 10, 100)
y = x
figure, ax = plt.subplots(2,2, figsize=(11,11))
ax[0][0].plot(x, y**k, 'r')
ax[0][1].plot(x, y**(k+1), 'b')
ax[0][1].plot(x, y**(k+2), 'r')
ax[1,0].plot(x, y**(k+2), 'b') #ax[1][0]
ax[1,1].plot(x, y**(k+3), 'b')
plt.show()
```

```python
# Para añadir leyendas utilizamos legend()
x = np.linspace(-2, 2, 100)
y = x ** 3
z = x ** 2
figure, ax = plt.subplots() # subplots generava la figura con sus dimensiones y los subgraficos
ax.plot(x, z, label="Función cuadrada", color="magenta")  
ax.plot(x, y, label="Función cubica")  
ax.legend()
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
# plt.xlabel("dhdhd")
# figure.savefig("hola.jpg") # guardar la imagen
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8)) # creamos el plot y sus caracteristicas generales

etiquetas = ["a","b","c","d"] # solo hay 1 vector de etiquetas
valores = [10,20,30,70] # azules
valores1 = [20,40,50,60] # rojos

# el vector de las x (su posicion numerica en el grafico)
x = np.arange(len(etiquetas))  # [0,1,2,3]

ancho = 0.5

# x-0.2 es donde esta el centro de la barra -0.2 (ancho=0.4) [-0.4,0]
plt.bar(x-0.25, valores, ancho, color="blue", alpha=0.3)
# x+0.2 es donde esta el centro de la barra
plt.bar(x + 0.25, valores1, width = ancho ,color="red", alpha=0.3) # 0.2 (ancho=0.4) -> [0,0.4]

ax.set_xticks(x) # donde pongo las etiquetas
ax.set_xticklabels(etiquetas)
plt.show()
```
