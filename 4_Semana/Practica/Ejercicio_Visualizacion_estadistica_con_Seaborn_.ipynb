# 1. Longitud del sépalo

```python
var = "sepal_l"
var_titulo = "Longitud del sépalo"

#creo subgráficos con seaborn
fig, ax = plt.subplots (1,3, figsize=(15,3))

#Boxplot
sns.boxplot(data=df[var], ax=ax[0], orient="h", width=0.4)
ax[0].set_xlabel(var_titulo)
ax[0].set_title("Boxplot de la longitud del sépalo")

#Histograma
sns.histplot(data=df[var], ax=ax[1])
ax[1].set_xlabel(var_titulo)
ax[1].set_ylabel("Frecuencia")
ax[1].set_title("Histograma de la longitud del sépalo")


#Curva de Densidad
sns.kdeplot(data=df[var], ax=ax[2], fill=True)
ax[2].set_xlabel(var_titulo)
ax[2].set_ylabel("Densidad")
ax[2].set_title("Densidad de la longitud del sépalo")

plt.show()
```

# Analisis Multivariante

## Longitud del sépalo vs Especies (setosa, virg, versicolor)

```python
var = "sepal_l"
var_titulo = "Longitud del sépalo"

#creo 2 subgráficos con seaborn (histograma, curva de densidad)
fig, ax = plt.subplots (1,2, figsize=(15,3), sharex=True)

#Histograma
sns.histplot(data=df_setosa, x=var, label="Setosa", ax=ax[0])
sns.histplot(data=df_versicolor, x=var, label="Versicolor", ax=ax[0])
sns.histplot(data=df_virginica, x=var, label="Virginica", ax=ax[0])
ax[0].set_xlabel(var_titulo)
ax[0].set_ylabel("Frecuencia")

#curva de densidad
sns.kdeplot(data=df_setosa, x=var, label="Setosa", fill=True, ax=ax[1])
sns.kdeplot(data=df_versicolor, x=var, label="Versicolor", fill=True, ax=ax[1])
sns.kdeplot(data=df_virginica, x=var, label="Virginica", fill=True, ax=ax[1])
ax[1].set_xlabel(var_titulo)
ax[1].set_ylabel("Densidad")
ax[1].legend()

plt.show()

```
