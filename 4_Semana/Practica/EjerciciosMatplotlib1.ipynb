## 1. Gr√°fica una recta de (-7,-7) a (7,7) de color rosa

```python
x = np.linspace(-7,7,20)
y = x

plt.plot(x,y,color="pink")
```

## 2. Obt√©n el siguiente resultado utilizando subplot y la ecuaci√≥n anterior

```python
x = np.linspace(-7,7,20)
y = x

for i, co in enumerate(["pink", "blue", "red", "magenta"]):
    plt.subplot(2,2,i+1)
    plt.plot(x,y,color=co)
```

## 3. Obt√©n la siguiente gr√°fica de barras 

```python
#Grafica de barras
etiquetas=["a","b", "c", "d"]
valores=[10, 20, 30, 40]
plt.bar(etiquetas, valores, color=["blue","yellow","cyan","magenta"])
plt.show()
```

## 4. Obtener la siguiente gr√°fica

```python
fig, ax = plt.subplots(figsize=(8,8)) 

letras = ["a","b","c","d"] 
cyans = [10,20,30,40] 
magentas = [20,30,40,50] 

x = np.arange(len(letras))  

ancho = 0.5

plt.bar(x-0.25, cyans, ancho, color="cyan")

plt.bar(x + 0.25, magentas, width = ancho ,color="magenta", alpha=1) 

ax.set_xticks(x) 
ax.set_xticklabels(letras)
plt.show()
```

## 7. Guarda el dataset de titanic en una variable

```python
titanic=sns.load_dataset("titanic")
titanic
```

## 8. Muestra en una grafica de barras el n√∫mero de hombres y mujeres que sobrevivieron

```python
supervivientes = titanic[titanic.survived==1].sex.value_counts()

plt.bar(supervivientes.index , supervivientes.values, color="green", width=0.5)
plt.show()
```

## 9. Muestra en una grafica de barras agrupada la cantidad de hombres y mujeres que sobrevivieron y la cantidad de los que no sobrevivieron

```python
fig, ax = plt.subplots(figsize=(6,6)) 

supervivientes = titanic[titanic.survived==1].sex.value_counts().sort_index()
no_supervivientes = titanic[titanic.survived==0].sex.value_counts().sort_index()

x = np.arange(2) 

plt.bar(x - 0.2 , supervivientes.values, color="blue", width=0.4)
plt.bar(x + 0.2, no_supervivientes.values, width = 0.4 ,color="red") 

ax.set_xticks(x) 
ax.set_xticklabels(supervivientes.index)
plt.show()
```

## 10. Muestra en una grafica de barras agrupada la cantidad de hombres y mujeres de cada clase (Pclass)

```python
fig, ax = plt.subplots(figsize=(5,5)) 

m1 = titanic[titanic["class"]=="First"].sex.value_counts().sort_index()
m2 = titanic[titanic["class"]=="Second"].sex.value_counts().sort_index()
m3 = titanic[titanic["class"]=="Third"].sex.value_counts().sort_index()

x = np.arange(2) 

plt.bar(x - 0.2 , m1.values, color="blue", width=0.2)
plt.bar(x, m2.values, width = 0.2 ,color="red") 
plt.bar(x + 0.2, m3.values, width = 0.2,color="green") 

ax.set_xticks(x) 
ax.set_xticklabels(m1.index)
plt.show()
```

## 11. Muestra en una gr√°fica el n√∫mero de sobrevivientes por clase (Pclass)

```python
fig, ax = plt.subplots(figsize=(5,3)) 

super_clase = titanic[titanic.survived==1]["class"].value_counts().sort_index()

x = np.arange(titanic["class"].unique().size) 

plt.bar(super_clase.index , super_clase.values, width=0.5, color=["blue","green","red"])
plt.show()
```

## 12. Muestra en una gr√°fica lineal las edades de los pasajeros que sobrevivieron y en otra la de los pasajeros que no sobrevivieron

```python
fig, ax = plt.subplots(figsize=(20,3)) 

y = titanic[titanic["survived"]==1].age
z = titanic[titanic["survived"]==0].age
ax.plot(y, label="Supervivientes", color="green")  
ax.plot(z, label="Fallecidos", color="red")  
ax.legend()
ax.set_xlabel("Index")
ax.set_ylabel("Edad")
plt.show()
```

## 13. Muestra en una grafica de caja la edad de los que sobrevivieron y de los que no sobrevivieron


Nota: Para poder graficarlo hay que reemplazar los NaN por el promedio de edad. 

```python
titanic.head(1)
```

```python
edadmedia = titanic.age.mean(skipna=True)
titanic.age.replace(np.nan, int(edadmedia),  inplace=True)
```

```python
edad = titanic.columns.get_loc("age")
supervi = titanic[titanic["survived"]==1].iloc[:,edad]
no_supervi = titanic[titanic["survived"]==0].iloc[:,edad]

sup_edad = [ no_supervi, supervi ]
titulos = ["Fallecidos", "Supervivientes"]

fig, ax = plt.subplots(figsize=(4,4))
ax.boxplot(sup_edad, labels=titulos)
plt.show()
```

## 14. Mostrar en una gr√°fica de caja por clase la edad de los pasajeros

```python
edad = titanic.columns.get_loc("age")
c1 = titanic[titanic["class"]=="First"].iloc[:,edad]
c2 = titanic[titanic["class"]=="Second"].iloc[:,edad]
c3 = titanic[titanic["class"]=="Third"].iloc[:,edad]

class_edad = [ c1,c2,c3 ]
titulos2 = ["Clase 1", "Clase 2", "Clase 3"]

fig, ax = plt.subplots(figsize=(4,4))
ax.boxplot(class_edad, labels=titulos2)
plt.show()
```
