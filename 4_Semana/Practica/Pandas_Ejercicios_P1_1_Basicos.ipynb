### 5. Operaciones por columnas:

c. Crea la variable topic_country concatenando las columnas topic y country

d. Ejecuta y entiende las siguientes líneas de código:

e. Elimina definitivamente del dataframe las columnas rate2, topic_country y country_num.

f. Sustituye los valores de AdWords, Reddit y SEO de la columna source por sus iniciales A, R y S.

g. Crea la variable muestra y que esta tome valor 1 para los valores de randUni menores a 0.7 y 0 para el resto.

h. Crea la variable submuestra y que esta tome valor 1 para los valores que muestra = 1 y que source = S.

```python
# a)
datos["rate2"] = datos.rate**(1/2)
datos
```

```python
# b)
datos["randUni"] = np.random.uniform(size=datos.shape[0])
datos.head()
```

```python
# c)
datos["topic_country"] = datos.topic + "_" + datos.country
datos
```

d) Ejecuta y entiende las siguientes líneas de código:

    datos["country_num"] = datos.country.str.slice(8,9).astype(int)
    crea "country_num", y le asigana el numero de country, como tipo int

    datos["country"] = "pais_" + datos["country_num"].astype(str)
    reescribe la columna countri pasandle como par="pais_", y el numeor de antes


```python
datos["country_num"] = datos.country.str.slice(8,9).astype(int)
datos["country"] = "pais_" + datos["country_num"].astype(str)
```

```python
# e) rate2, topic_country y country_num.
del datos["rate2"]
del datos["topic_country"]
del datos["country_num"]
datos
```

```python
# f) Sustituye los valores de AdWords, Reddit y SEO de la columna source 
#    por sus iniciales A, R y S.
datos.source = datos.source.replace(["AdWords", "Reddit", "SEO"],["A","R","S"])
```

```python
# g)Crea la variable muestra y que esta tome valor 1 para los valores de randUni menores a 0.7 y 0 para el resto.
datos["muestra"] = 0
datos.muestra[datos.randUni < 0.7] = 1
datos

# csv.loc[csv["Largo Sepalo"]>=5, "Largo Sepalo"] = 5.5
```

```python
# h)Crea la variable submuestra y que esta tome valor 1 para los valores que muestra = 1 y que source = S.
datos["submuestra"] = 0
datos.submuestra[(datos.muestra == 1) & (datos.source == "S")] = 1
datos
```
