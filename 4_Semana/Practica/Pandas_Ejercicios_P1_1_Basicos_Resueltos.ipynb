### 5. Operaciones por columnas:
a. Crea la variable rate2 como la raíz cuadrada de la variable rate.

b. Crea la variable randUni a partir de generar valores aleatorios de una distribución uniforme[0,1]

c. Crea la variable topic_country concatenando las columnas topic y country

d. Ejecuta y entiende las siguientes líneas de código:
datos["country_num"] = datos.country.str.slice(8, 9).astype(int)
datos["country"] = "pais_" + datos["country_num"].astype(str)
e. Elimina definitivamente del dataframe las columnas rate2, topic_country y country_num.
f. Sustituye los valores de AdWords, Reddit y SEO de la columna source por sus iniciales A, R y S. 

g. Crea la variable muestra y que esta tome valor 1 para los valores de randUni menores a 0.7 y 0 para el resto.

h. Crea la variable submuestra y que esta tome valor 1 para los valores que muestra = 1 y que source = S.

```python
import numpy as np
datos["rate2"] = datos["rate"]**(1/2)
datos["randUni"] = np.random.uniform(0, 1, datos.shape[0])
datos["topic_country"] = datos["topic"] + datos["country"]
datos["country_num"] = datos.country.str.slice(8, 9).astype(int)
datos["country"] = "pais_" + datos["country_num"].astype(str)
datos.drop(columns = ["rate2", "topic_country", "country_num"], inplace = True)
datos.replace(['AdWords', 'Reddit', 'SEO'], ["A", "R", "S"], inplace=True)

```

```python
datos.loc[datos["randUni"] < 0.7,'muestra']=1
datos.loc[datos["randUni"] >= 0.7,'muestra']=0
datos.loc[(datos["randUni"] < 0.7) & (datos["source"] == "S"),'submuestra']=1
datos.loc[~((datos["randUni"] < 0.7) & (datos["source"] == "S")),'submuestra']=0
```

```python
# también se puede hacer con el np.where
datos['muestra'] = np.where(datos["randUni"] < 0.7, 1, 0)
datos['submuestra'] = np.where((datos["randUni"] < 0.7) & (datos["source"] == "S"), 1, 0)
```
