### 26. Obtener el promedio y la mediana de cada año de migración y agregar al dataframe anterior

```python
df = pd.read_excel("Canada.xlsx")
df.loc[:,1980:].max()
df3 = pd.DataFrame(list(df.loc[:,1980:].max().to_dict().items()), columns=['Año', 'N_Migrantes'])
df3.head(5)
```

```python
df2 = df.loc[:,1980:].T
df2 = df2.agg([np.mean, np.median], axis=1)
df2 = df2.reset_index()
df2 = df2.rename(columns={"index": "Año"})
df4 = df3.merge(right=df2, on="Año" )
df4
```
