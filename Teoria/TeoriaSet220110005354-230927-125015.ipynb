<font size="7"><span style="color: #33D6CC;">&iquest;Qu&eacute; es un set?</span></font>

<font size="3"><span style="color: #5E62D1;">En Python, <strong> un set </strong> es un objeto similar a las listas o los diccionarios, que vers a continuacin. Se crean usando llaves <strong> {} </strong> y <strong> no se pueden indexar </strong>. Pero <strong> son mutables </strong> ya que se pueden agregar o eliminar elementos.</span></font>

<font size="3"><span style="color: #5E62D1;"> Por la manera en la que se almacenan, es <strong> ms rpido </strong> aadir un elemento o comprobar si est o no en el set.</span></font>

<font size="3"><span style="color: #5E62D1;"> Pueden estar formados por objetos de cualquie tipo: string, int, float, boolean...</span></font>

<font size="3"><span style="color: #5E62D1;"> Los <strong> sets </strong> presentan una propiedad muy importante <strong> no contienen elementos duplicados </strong> que se va a respetar siempre que creemos un set.</span></font>

<font size="3"><span style="color: #5E62D1;">Son una estructura realmente til, pues van a tener mtodos que nos permiten realizar las operaciones ms frecuentes de la <strong> teora de conjuntos </strong></span></font>

<font size="3"><span style="color: #5E62D1;">Veamos algunos ejemplos de uso:</span></font>

```python
mi_set = {1, 'a', False, 'Hola Mundo'}
print(mi_set) #Podemos combinar valores de distintos tipos
```

```python
mi_set = {1, 5, 2, 1, 3, 1, 4, 5, 6}
print(mi_set) #No va a almacenar los elementos duplicados
print(len(mi_set)) #Podemos usar la funcin len() para devolver el nmero de elementos del set
```

<font size="3"><span style="color: #5E62D1;"> Algunos <strong> mtodos </strong> que nos permiten realizar las operaciones ms habituales de <strong> la teora de conjuntos </strong> son: </span></font>



<ul>
<font size="3"><li><span style="color: #5E62D1;"><strong> .union()</strong> o tambin <strong>"|" </strong>: realiza la unin de dos conjuntos y devuelve un nuevo conjunto.</span></li></font>
</ul>

```python
set1 = {0,1,2,3,4}
set2 = {2,3,4,5,6}
print(set1.union(set2)) #Devuelve un nuevo set con la unin de los dos anteriores
```

```python
set3 = set1 | set2 
set3
```

<font size="3"><li><span style="color: #5E62D1;"><strong> .intersection()</strong> o tambin <strong>"&"</strong>: realiza la interseccin de dos conjuntos y devuelve un nuevo conjunto.</span></li></font>


```python
set1 = {0,1,2,3,4}
set2 = {2,3,4,5,6}
print(set1.intersection(set2)) #Devuelve un nuevo set con la interseccin
```

```python
set1 & set2
```

<font size="3"><li><span style="color: #5E62D1;"><strong> .difference() </strong> o tambin <strong> "-" </strong>: realiza la diferencia de dos conjuntos, es decir, devuelve un nuevo conjunto con los elementos que no estn en el segundo conjunto.</span></li></font>

```python
set1 = {0,1,2,3,4}
set2 = {2,3,4,5,6}
print(set1.difference(set2)) #Devuelve un nuevo set con los elemenos de set1 que no estn en set2
print(set2.difference(set1)) #Devuelve un nuevo set con los elemenos de set2 que no estn en set1
```

```python
set1 - set2
```

<font size="3"><span style="color: #5E62D1;"> El mtodo <strong> .update() </strong> actualiza un set con los valores del set (o iterable) introducido como parmetro.</span></font>

<font size="3"><span style="color: #5E62D1;"> Es muy parecido al mtodo .union(), pero en su lugar, modifica el valor del set, es decir, no crea un set nuevo.</span></font>

```python
mi_set = {1,2,3,4}
set3 = {98,99}
mi_set.update(set3)
print(mi_set)
```

```python
mi_set = {1,2,3,4}
set3 = {98,99}
mi_set.union(set3)
print(mi_set) #Vemos como el mtodo union NO modifica el valor de mi_set
```
