```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
train=pd.read_csv("train.csv")
```

```python
train.head()
```

```python
# Tiro la columna id, course en train
train=train.drop(columns=['id','Course'])
```

```python
X=train.drop(columns=['Target'])
Y=train['Target']
```

## El test de kaggle no me sirve para evaluar mi modelo(no tiene Y), necesito apartar unos datos del train para poder evaluar el modelo. (HarÃ© mi propio test)

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
```

```python
# hago el preprocesamiento en el X_train
scaler=StandardScaler()
scaler.fit(X_train)
X_train[:]=scaler.transform(X_train)
X_test[:]=scaler.transform(X_test)
```

```python
clf=LogisticRegression(max_iter=10000)
clf.fit(X_train,y_train)
```

```python
yhat_train=clf.predict(X_train)
accuracy_score(y_train,yhat_train)
```

```python
yhat_test=clf.predict(X_test)
accuracy_score(y_test,yhat_test)
```
