<font size="7"><span style="color: #33D6CC;">&iquest;Qu&eacute; es una tupla?</span></font>

<font size="3"><span style="color: #5E62D1;">Una tupla al igual que las listas, se utiliza para **almacenar un conjunto de elementos secuenciados de cualquier tipo** y estos elementos tambin tienen un **ndice** como en el caso de las listas o cadenas.</span></font>

<font size="3"><span style="color: #5E62D1;">La principal diferencia es que las tuplas son **inmutables**, sus valores no se pueden eliminar de la tupla.</span></font>

<font size="6"><span style="color: #33D6CC;">Funcin ZIP</span></font>

<font size="3"><span style="color: #5E62D1;"> A partir de dos secuencias (colecciones, listas, tuplas..., la funcin ZIP entrelaza dichos elementos creando un objeto iterable como se muestra a continuacin.</span></font>


```python
lista = ["a", "b", "c",4]
tupla = (5,6,7,5)

objeto_zip = zip(lista,tupla)
print(objeto_zip)
```

```python
x = 10
print(x)
```

```python
for i in objeto_zip:
    print(i)
```

```python
for i in zip(lista,tupla):
    print(i)
```
