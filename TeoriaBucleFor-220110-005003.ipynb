<font size="7"><span style="color: #33D6CC;">&iquest;Qu&eacute; es un bucle for?</span></font>


<font size="3"><span style="color: #5E62D1;">El <strong>bucle for</strong> nos permite <strong>iterar</strong> un bloque de cdigo en una secuencia, a lo largo de una lista, de un string, de una tupla o incluso de un diccionario.</span></font>



```python
for i in [4,5,6,7]:
    print(i)
```

<font size="3"><span style="color: #5E62D1;">Veamos <strong>un ejemplo</strong> en el que vamos a imprimir las palabras de una lista aadiendo despus de cada una de ellas un @.</span></font>

```python
lista_palabras = ['hola', 'mundo', 'adios', 'python'] #Declaramos una lista de palabras
for palabra in lista_palabras: #definimos una variable que va a iterar en toda la lista, cuyo nombre ser "palabra"
    print(palabra + "@") #IMPORTANTE: el cdigo dentro de un bucle for, ir indentado con un tabulador
```

<font size="3"><span style="color: #5E62D1;">En el cdigo de arriba, la <strong>variable palabra </strong> hace referencia a cada uno de los <strong>elementos de la lista</strong> en cada iteracin del bucle for.</span></font>

<font size="3"><span style="color: #5E62D1;">Su valor en la primera iteracin ser igual a "hola" y en la ltima ser "python"</span></font>

```python
lista_palabras = ['hola', 'mundo', 'adios', 'python'] #Declaramos una lista de palabras
for i in [0, 1, 2, 3]: #Recorremos la secuencia de 4 valores (0,1,2,3) generado por la funcin range()
    print(lista_palabras[i] + "@") #Accedemos a los elemnos de la lista por su ndice
```

<font size="3"><span style="color: #5E62D1;">En este caso, la variable "i" es la que va ir tomando los diferentes valores de la secuencia de forma ordenada. Usaremos esto, para acceder a los elementos de la lista de palabras.</span></font>

<font size="3"><li><span style="color: #5E62D1;">Si queremos recorrer los valores de una lista y adems, mantener el ndice en una variable, podemos utlizar la funcin <strong> enumerate()</strong>.</span></li></font>

<font size="3"><li><span style="color: #5E62D1;"><strong> enumerate()</strong> va a devolver en cada iteracin una tupla cuyo primer elemento es el ndice de la lista y el segundo el propio valor de la lista en esa posicin. Podemos utilizar las propiedades de las tuplas para definir en un mismo bucle las variables "indice" y "animal".</span></li></font>

```python
lista_animales = ['Perro', 'Gato', 'Perico'] 
for animal in enumerate(lista_animales):
    print(animal)
```

```python
lista_animales = ['Perro', 'Gato', 'Perico',"Loro"] 
for indice, animal in enumerate(lista_animales): #Utilizamos el unpack de tuplas
    print(indice,animal)
```

<font size="3"><li><span style="color: #5E62D1;">Es posible <strong> anidar</strong> varios bucles for, esto significa que en una iteracin, se realizarn varias iteraciones.</span></li></font>

```python
lista_animales = ['perro', 'gato', 'perico']
lista_adjetivos = ['feo', 'bonito']
for animal in lista_animales: #Se realizarn 3 iteraciones, una para cada animal
    for adj in lista_adjetivos: #Para cada animal se realizarn dos iteraciones sobre la lista de adjetivos
        print('El ' + animal + ' es ' + adj)
```
