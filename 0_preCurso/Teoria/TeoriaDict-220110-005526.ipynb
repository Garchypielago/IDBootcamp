<font size="7"><span style="color: #33D6CC;">&iquest;Qu&eacute; es un diccionario?</span></font>

<font size="3"><span style="color: #5E62D1;"> Los <strong> diccionarios </strong> en Python son otro objeto bsico que nos permite asignar a ciertas <strong> claves (keys) </strong> determinados <strong> valores (values). </strong> Se declaran como los sets, mediante llaves <strong> {} </strong>.</span></font>

<font size="3"><span style="color: #5E62D1;"> Como en el caso de las listas, tienen indexacin, pero con una peculiaridad. Los ndices sern lo que hemos llamado <strong> claves (keys) </strong>. Cada elemento de un diccionario est representado como un par<strong> clave:valor (key:value) </strong>. A este par de clave, valor, se le llama <strong> item. </strong> </span></font>

```python
mi_dict = {5:32 , 'clave2':'Hola Mundo'}
#Esto es un ejemplo de declaracin de un diccionario con dos claves y diferentes valores (32 y 'Hola Mundo')
```

```python
mi_dict
```

```python
print(mi_dict[5]) #Imprime el valor correspondiente a la clave "clave1", que es 32
```

<font size="3"><span style="color: #5E62D1;"> Como <strong> clave </strong> en un diccionario solo es posible utilizar <strong> objetos inmutables </strong>, que son aquellos que no pueden cambiar de valor. Ya conoces algunos de ellos como <strong> int, str, float, bool...</strong></span></font>

```python
error_dict = {"lista": [1,2,3,4]} #Intentamos declarar un diccionario con una clave de tipo mutable (list)
```

```python
error_dict
```

<font size="3"><span style="color: #5E62D1;"> Por defecto, si utilizamos el operador <strong> in </strong> en un diccionario, estamos comprobando la condicin <strong> dentro de las claves.</strong> Veamos un ejemplo: </span></font>

```python
numeros = {1: 'uno', 2:'dos', 3:'tres'}
print('uno' in numeros) #'uno' no es una clave de nuestro diccionario, por lo tanto, ser False
print(1 in numeros) # 1 s es una clave de nuestro diccionario, por lo tanto, ser True
```

<font size="3"><span style="color: #5E62D1;"> Un <strong> mtodo </strong> muy til en los diccionarios es <strong> .get()</strong> que nos devuelve el valor asociado a la clave que utilicemos.</span></font>

```python
mi_dict = {'clave':3, '1':False, 2:1}
print(mi_dict.get('clave')) #Imprimimos el valor asociado a la clave "clave"
```

<font size="3"><span style="color: #5E62D1;"> Podemos pasar al mtodo get, un parmetro extra que evitar que nos devuelva un valor <strong> None </strong> si la clave no existe (devolver este valor). </span></font>

```python
mi_dict = {'clave':3, '1':False, 2:1}
print(mi_dict.get('hola')) #Nos devuelve None porque 'hola' no es una clave de nuestro diccionario
print(mi_dict.get('hola', 'Hola mundo')) #Si 'hola' no es una clave, devuelve "No existe la clave"
```

<font size="3"><span style="color: #5E62D1;"> Como en el resto de objetos, podemos usar la funcin <strong> len() </strong> para ver el nmero de elemenos (items) del diccionario. </span></font>

```python
mi_dict = {'1':1, '2':2, '3':3}
print(len(mi_dict))
```

<font size="3"><span style="color: #5E62D1;"> Veamos algunos <strong> mtodos </strong> tiles para utilizar en diccionarios: </span></font>

<ul>
<font size="3"><li><span style="color: #5E62D1;"><strong> .values() </strong>: devuelve los valores del diccionario.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .keys() </strong>: devuelve las claves del diccionario.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .items() </strong>: nos devuelve los elementos del diccionario en forma de tuplas (clave, valor).</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .pop() </strong>: elimina un elemento del diccionario a partir de su clave.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .popitem() </strong>: elimina el ltimo elemento del diccionario.</span></li></font>
<font size="3"><li><span style="color: #5E62D1;"><strong> .update() </strong>: de manera similar al mismo mtodo en los sets, actualiza los valores del diccionario, modificando su valor.</span></li></font>
</ul>

```python
mi_dict = {'Adrian': 28, 'Alfonso': 34, 'Messi': 45, 'Turing': 56, 'Biden': 89}
print(mi_dict.values()) #Podemos observar que es un elemento de tipo "dict_values", podemos transformarlo a lista
```

```python
mi_dict = {'Adrian': 28, 'Alfonso': 34, 'Messi': 45, 'Turing': 56, 'Biden': 89}
print(mi_dict.values()) #Podemos observar que es un elemento de tipo "dict_values", podemos transformarlo a lista
print(list(mi_dict.values())) #usamos list() para transformar el objeto a tipo lista
```

```python
mi_dict = {'Adrian': 28, 'Alfonso': 34, 'Messi': 45, 'Turing': 56, 'Biden': 89}
print(mi_dict.keys()) #Mismo caso que con values
```

```python
print(list(mi_dict.keys())) #Mismo caso que con values
```

```python
mi_dict = {'Adrian': 28, 'Alfonso': 34, 'Messi': 45, 'Turing': 56, 'Biden': 89}
print(mi_dict.items()) #Nota: si lo convertimos a lista, tendremos una lista de tuplas
```

```python
print(list(mi_dict.items())) #Nota: si lo convertimos a lista, tendremos una lista de tuplas
```

```python
mi_dict = {'Adrian': 28, 'Alfonso': 34, 'Messi': 45, 'Turing': 56, 'Biden': 89}
print(list(mi_dict.items())) #Nota: si lo convertimos a lista, tendremos una lista de tuplas
```

```python
mi_dict = {'Adrian': 28, 'Alfonso': 34, 'Messi': 45, 'Turing': 56, 'Biden': 89}
borrar = mi_dict.pop('Messi') #.pop() devuelve el valor eliminado
print(mi_dict) #Hemos eliminado a Messi y hemos guardado su valor en la variable "borrar"
print(borrar) #Imprime 45 que es el valor asociado a Messi
```

```python
mi_dict = {'Adrian': 28, 'Alfonso': 34, 'Messi': 45, 'Turing': 56, 'Biden': 89}
borrar_popitem = mi_dict.popitem() #.popitem() devuelve el valor eliminado
print(mi_dict) #Hemos eliminado el ltimo elemento del diccionario
print(borrar_popitem) #NOTA: Imprime el ltimo elemento del diccionario (clave y valor)
```

```python
mi_dict = {'Adrian': 28, 'Alfonso': 34, 'Messi': 45, 'Turing': 56, 'Biden': 89}
mi_dict.update({'Biden': 90}) #Podemos utilizar el mtodo .update() para aadir de nuevo el par "Messi:60" al diccionario
print(mi_dict)
```

```python
mi_dict = {'Adrian': 28, 'Alfonso': 34, 'Messi': 45, 'Turing': 56, 'Biden': 89}
mi_dict['Turing2'] = 60 #Tambin podemos aadir pares de clave:valor al diccionario de manera explicita
print(mi_dict)
```
