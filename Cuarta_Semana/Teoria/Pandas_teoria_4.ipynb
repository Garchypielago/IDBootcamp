# groupby -> transform

```python
import pandas as pd

df = pd.DataFrame({'a':[1,2,3,4,5,6],
               'b':[7,8,9,10,11,12],
               'c':['q', 'q', 'w', 'q', 'q', 'w']
              })

df
```

```python
df.groupby("c").mean()
```

```python
df.groupby("c")[['a','b']].transform('mean')
```

```python
df[["mean_a","mean_b"]]=df.groupby("c").transform("mean")
df
```

```python
df = pd.read_excel("Canada.xlsx")
df.head(2)
```

```python
df.groupby(["Continente"])[[1980]].mean() #Devuelve el tamaño de cada grupo de la agrupación
```

```python
df.groupby(["Continente"])[[1980]].transform('mean')
```

```python
df["media_1980_continente"]=df.groupby(["Continente"])[[1980]].transform('mean')
```

```python
df
```

## aggregate a un mismo groupby puedo aplicarle varias funcones de agrupado

```python
f=lambda x: np.sum(x**2)
# apply => una columna, se aplica a cada valor por separado
# con el agg, puedo aplicar una función externa, pero considerando que stiene que ser una funcion de grupo
df_group = df.groupby("Continente")[[2010,2011,2012]].agg(["min", "max", f, np.mean])
```

```python
df_group
```

```python
df_group=df.groupby("Continente")[list(range(1980,1988))].aggregate([min, max, sum ])
df_group
```

```python
df.groupby("Continente")[[2010,2011,2012]].agg(["min", "max", f, np.mean])
```

```python
df.groupby("Continente", as_index=False)[[2010,2011,2012]].mean()
# con as_index=False, no lo pone como indice
```

```python
df5 = df.groupby("Continente").agg({2010:["min",'max','sum'],
                                    2011:['min','max'],
                                    2012:["sum"]})
df5
```

```python
# para acceder a min de 2010, primero accedo a la serie 2010
df5[2010]
# despues accedo a la columna del min
df5[2010]['min']
```

## Melt
#### Puedo seleccionar columnas con un range o un np.arange

```python
df.head()
```

```python
df_final=df[["Pais",*range(1980,2010)]]
# asi puedo seleccionar las columnas pais y del 1980 a 2010
```

```python
df_final.head()
```

```python
df_nuevo=pd.melt(df, 
                 id_vars=["Pais"], # esta columna no va a cambiar, seguirá existiendo
                 value_vars=list(range(1980,2010)),  # aqui estan los nombres de las columnas desde donde cogeremos los valores
                 var_name="Año", # Nombre que tendrá la columna que tiene como valores el nombre de las columnas antiguas
                 value_name= "Valor") # Nombre que tendrá la columna que tiene como valoreslos valores antiguos
```

```python
df_nuevo
```

```python
df_nuevo.sort_values("Valor", ascending=False)
```

```python
df_nuevo[df_nuevo.Pais=="Spain"]
```

# Pivot: proceso contrario a melt 

```python
data=pd.pivot(df_nuevo, # dataframe como el resultado de un melt
              index="Pais", # la variable que tomara como indice de las filas
              columns="Año", # los valores de esta columna (año) seran los nombres de las columnas del dataframe resultante
              values="Valor") # los valores de esta columna (año) seran los valores del dataframe resultante 
```

```python
data
```
