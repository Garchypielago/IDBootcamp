```python
import pandas as pd

df = pd.DataFrame({'a':[1,2,3,4,5,6],
               'b':[7,8,9,10,11,12],
               'c':['q', 'q', 'q', 'q', 'w', 'w']
              })

df
```

```python
df.groupby("c").mean()
```

```python
df.groupby("c").transform("mean")
```

```python
df[["suma_a","suma_b"]]=df.groupby("c").transform(np.sum)
df
```

```python
df = pd.read_excel("Canada.xlsx")
```

```python
df.groupby(["Continente"])[[1980]].mean() #Devuelve el tamaño de cada grupo de la agrupación
```

```python
df["media_1980_continente"]=df.groupby(["Continente"])[[1980]].transform('mean') # Cuenta el número de elementos que no son NaNs
```

```python
# dataframe.groupby(columnas)[[lista de columnas que quiero en el resultado]].funcion()
df.groupby("Continente")[[1980,1990]].mean() #Agrupamos por continente y luego sumamos la migración de cada continente en el año 1980
```

```python
df[df.Region>100].groupby("Continente")[np.arange(2008,2012)].sum()
# filtrando el df y aplicando un groupby a ese filtrado
```
