# Crear dataframe desde 0

```python
import pandas as pd
import numpy as np


listas = [['Gael', 25], ['Paola', 30], ['Maria', 26], ['Jesus', 22]]
# desde una lista de listas indicando el nombre de las columnas
df = pd.DataFrame(listas, columns =['Nombre', 'Edad'])
df
```

#### .isin(lista de elementos)

```python
lista_ciudades = ["Madrid","Bilbao"]
df["Residencia"].isin(lista_ciudades)
df[df["Residencia"].isin(["Madrid","Bilbao"])]
df[df.Residencia.isin(["Madrid","Bilbao"])]
```

#### between(min, max)

```python
df
```

```python
df["Edad"].between(28,30)
```

```python
#between para elegir filas entre 29 y 30 años
df[df["Edad"].between(28,30)] # filtraje por intervalo numerico
```

```python
(df["Edad"].between(28,30))
```

#### ~ == not (negación de la condición)

```python
~(df["Edad"].between(28,29))
# virgulilla = alt_gr+4
```

```python
#between y virgulilla para elegir elementos que no esten entre 29 y 30
df[~df["Edad"].between(28,29)] #Tenemos el filtro lo ponemos entre parentesis y anteriormente se pone la virgulilla
```

## Merge

```python
df1 = pd.DataFrame({"Nombre":["Gael","Paola","Adrian"], "id":[1,2,3], "edad":[30,31,23]})
df1
```

```python
df2 = pd.DataFrame({"id":[1,1,1,2,2,2,2,4,4], 
                    "Producto":["a","b","a","b","c","c","d","d","f"]})
df2
```

```python
pd.merge(left=df1, right=df2, on = "id")
# sii no pongo nada es un how="inner"=> solo mantiene los id que estan en los 2 dataframes
```
