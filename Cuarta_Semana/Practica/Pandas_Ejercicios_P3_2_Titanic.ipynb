**18. Agrega una nueva columna al daframe que se llame: "Rango edad" y va a catagorizar de la siguiente manera:**
- Niño < 11 años
- Joven <18 años
- Adulto < 50 años
- Mayor >= 50 años

```python
def rango(x):
    if x<11:
        return "Niño"
    if x<18:
        return "Joven"
    if x<50:
        return "Adulto"
    else:
        return "Mayor"
```

```python
df["RangoEdad"]=df.Age.apply(rango)
df
```
